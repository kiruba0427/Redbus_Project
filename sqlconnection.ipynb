```python
#importing libraries
import pandas as pd
import mysql.connector
import numpy as np
```

```python
# csv to dataframe
df_buses_kerala=pd.read_csv(r'C:\Users\kirub\OneDrive\Desktop\Redbus\df_buses_kerala.csv')
df_buses_Andhra=pd.read_csv(r'C:\Users\kirub\OneDrive\Desktop\Redbus\df_buses_Andhra.csv')
df_buses_Telugana=pd.read_csv(r'C:\Users\kirub\OneDrive\Desktop\Redbus\df_buses_Telugana.csv')
df_buses_Goa=pd.read_csv(r'C:\Users\kirub\OneDrive\Desktop\Redbus\df_buses_Goa.csv')
df_buses_Rajasthan=pd.read_csv(r'C:\Users\kirub\OneDrive\Desktop\Redbus\df_buses_Rajasthan.csv')
df_buses_SouthBengal=pd.read_csv(r'C:\Users\kirub\OneDrive\Desktop\Redbus\df_buses_SouthBengal.csv')
df_buses_Haryana=pd.read_csv(r'C:\Users\kirub\OneDrive\Desktop\Redbus\df_buses_Haryana.csv')
df_buses_Assam=pd.read_csv(r'C:\Users\kirub\OneDrive\Desktop\Redbus\df_buses_Assam.csv')
df_buses_UttarPradesh=pd.read_csv(r'C:\Users\kirub\OneDrive\Desktop\Redbus\df_buses_UttarPradesh.csv')
df_buses_WestBengal=pd.read_csv(r'C:\Users\kirub\OneDrive\Desktop\Redbus\df_buses_WestBengal.csv')

Final_df=pd.concat([df_buses_kerala,df_buses_Andhra,df_buses_Telugana,df_buses_Goa,df_buses_Rajasthan,df_buses_SouthBengal,
                    df_buses_Haryana,df_buses_Assam,df_buses_UttarPradesh,df_buses_WestBengal],ignore_index=True)
Final_df
```

```python
# data about the data
Final_df.info()
```

```python
print(Final_df["Price"].dtype)
Final_df["Price"] = Final_df["Price"].astype(str)
Final_df["Price"] = Final_df["Price"].str.replace("INR", "")
Final_df["Price"] = pd.to_numeric(Final_df["Price"], errors="coerce")
Final_df["Price"].fillna(0, inplace=True)
```

```python
Final_df["Price"]
```

```python
#convert Ratings to numeric
Final_df["Ratings"]=Final_df["Ratings"].str.replace("New","")
Final_df["Ratings"]=Final_df["Ratings"].str.strip()
Final_df["Ratings"]=Final_df["Ratings"].str.split().str[0]
Final_df["Ratings"] = pd.to_numeric(Final_df["Ratings"], errors='coerce')
Final_df["Ratings"].fillna(0,inplace=True) 
```

```python
Final_df["Ratings"]
```

```python
# info after the data type change
Final_df.info()
```

```python
Final_df = Final_df[Final_df["Price"] <= 7000]
```

```python
# replacing the nan value
Final_df = Final_df.replace({np.nan: None})
```

```python
Final_df["Bus_type"]
```

```python
# Option 1: Fill missing values in Bus_type (replace None with a placeholder like 'Unknown')
Final_df['Bus_type'].fillna('Unknown', inplace=True)
```

```python
# Check for missing values in Seats_Available
missing_seats = Final_df[Final_df['Seats_Available'].isnull()]
print(missing_seats)
```

```python
Final_df['Seats_Available'].fillna('0', inplace=True)  # Replace NaN with '0'
```

```python
Final_df['Seats_Available'].replace('', '0', inplace=True)  # Replace empty strings with '0'
```

```python
Final_df.dropna(subset=['Start_time'], inplace=True)  # Remove rows with NaN in Start_time
```

```python
# change dataframe to csv
path="C:/Users/kirub/OneDrive/Desktop/Redbus/Final_df.csv"
Final_df.to_csv(path,index=False)
```

```python
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="Kiru@2000",
    database="RED_BUS_DETAILS"
)
my_cursor = conn.cursor()
```

```python
#'''The ID column will increment automatically each time a new row is added, providing a unique, 
#sequential identifier for each bus record. This approach keeps the data organized, ensures uniqueness, 
#and simplifies data management within the bus_details table.'''


# Table Creation
my_cursor.execute('''CREATE TABLE IF NOT EXISTS bus_details(
                  ID INT AUTO_INCREMENT PRIMARY KEY,
                  Bus_name VARCHAR(255) NOT NULL,
                  Bus_type VARCHAR(255) NOT NULL,
                  Start_time VARCHAR(255) NOT NULL,
                  End_time VARCHAR(255) NOT NULL,
                  Total_duration VARCHAR(255) NOT NULL,
                  Price FLOAT NULL,
                  Seats_Available VARCHAR(255) NOT NULL,
                  Ratings Float NULL,
                  Route_link VARCHAR(255) NULL,
                  Route_name VARCHAR(255) NULL
                  )''')
print("Table Created successfully")
```

```python
# SQL query to insert data into bus_details table
insert_query = '''INSERT INTO bus_details(
                    Bus_name,
                    Bus_type,
                    Start_time,
                    End_time,
                    Total_duration,
                    Price,
                    Seats_Available,
                    Ratings,
                    Route_link,
                    Route_name)
                  VALUES(%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)'''
data = Final_df.values.tolist()

my_cursor.executemany(insert_query, data)

conn.commit()

print("Values inserted successfully")
```
